feature engineering

```python
def vehiclecategory(price):
    if price < 50000:
        return "Entry Level"
    elif 50000 <= price <100000:
        return "Budget Range"
    elif 100000 <= price < 250000:
        return "Mid Range"
    else:
        return "Premium Range"


```

```python
df['Price_Range'] = df['Price'].apply(vehiclecategory)
# df
df["Price_Range"].value_counts()
df[df["Price_Range"]=="Premium Range"]
```

```python
# Count the number of cars in each Price_Range category
price_range_counts=df['Price_Range'].value_counts()
plt.figure(figsize=(8, 5))
plt.bar(price_range_counts.index, price_range_counts.values, color=['lightblue', 'orange', 'green'])
plt.title('Sales Count by Price Range')
plt.xlabel('Price Range')
plt.ylabel('Number of Cars Sold')
plt.show()
```

```python
# Group by Color and Brand to count the number of vehicles for each combination
color_brand_count = df.groupby(['Color', 'Brand']).size().reset_index(name='Count')
most_sold_color = df['Color'].value_counts().idxmax()  # Get the most frequent color
most_sold_color_data = color_brand_count[color_brand_count['Color'] == most_sold_color]
most_sold_color_data_sorted = most_sold_color_data.sort_values(by='Count', ascending=False)
plt.figure(figsize=(10, 6))
plt.bar(most_sold_color_data_sorted['Brand'], most_sold_color_data_sorted['Count'], color='skyblue')
plt.title(f"Brands for the Most Sold Color: {most_sold_color}")
plt.xlabel("Brand")
plt.ylabel("Count of Vehicles")
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()

```

```python
#kilometers according to year
avg_kilometers_per_year = df.groupby('Year')['Kilometers'].mean()
plt.figure(figsize=(10, 6))
avg_kilometers_per_year.plot(kind='bar', color='skyblue')
plt.title("Average Kilometers Driven According to Year")
plt.xlabel("Year")
plt.ylabel("Average Kilometers Driven")
plt.tight_layout()
plt.show()
```

```python
# Sort the dataframe by Kilometers in descending order and get the top 5 vehicles
top_5_km_vehicles = df.sort_values(by='Kilometers', ascending=False).head(5)
plt.figure(figsize=(10, 6))
plt.scatter(top_5_km_vehicles['Name'], top_5_km_vehicles['Kilometers'],color='skyblue', s=100)
plt.title("Top 5 Vehicles with Highest Kilometers Driven")
plt.xlabel("Vehicle Name")
plt.ylabel("Kilometers Driven")
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()


```

```python
# Group by vehicle type and calculate average price
avg_price_by_type = df.groupby('Type')['Price'].mean()
plt.figure(figsize=(12, 6))
avg_price_by_type.sort_values(ascending=False).plot(kind='bar', color='purple')
plt.title('Average Price by Vehicle Type', fontsize=15)
plt.xlabel('Vehicle Type', fontsize=12)
plt.ylabel('Average Price', fontsize=12)
plt.show()
```
